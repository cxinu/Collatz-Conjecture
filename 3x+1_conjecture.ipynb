```python
import matplotlib.pyplot as plt
import math

num = int(input('x = '))
count, X, Y = -1, [], []

def C3xp1(x):
  """
  3x+1 conjecture states that,
  starting from any positive integer n,
  repeated iteration of this function eventually produces the value 1.
  """

  global count, X, Y
  count += 1
  X.append(count)
  Y.append(x)

  if x == 1:
    return 1
  elif x%2 == 0:
    return C3xp1(x/2)
  else:
    return C3xp1(3*x+1)

C3xp1(num)
print('Stopping time =', count)
```

```python
plt.title('Collatz conjecture')
plt.plot(X, Y, color='orange', linewidth = 3,
		marker='o', markerfacecolor='purple', markersize=7)
plt.show()
```

```python
plt.title('Logarithmic downward trend')
logY = list(map(math.log, Y))
plt.plot(X, logY, color='orange', linewidth = 3,
		marker='o', markerfacecolor='black', markersize=7)
plt.show()
```
